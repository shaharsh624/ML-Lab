## Concatenate both dataset

```python
fig = px.histogram(full_data, x='type', color_discrete_sequence=['deepskyblue'])

fig.update_layout(
    title='Distribution Of Type',
    title_font_size=20,
    title_font_color='black',
    xaxis_title='Type',
    yaxis_title='Count',
    xaxis_tickangle=-60,
    xaxis_showgrid=True,
    yaxis_showgrid=True
)

fig.show()
```

```python
fig = px.line(full_data, x='Year', y='Scheduled', title='Total Flights per Year', labels={'Scheduled': 'Total Flights (in Million)'}, color_discrete_sequence=['#ff7f0e'])
fig.add_scatter(x=full_data['Year'], y=full_data['Charter'], mode='lines', name='Chartered Flights', line=dict(color='#1f77b4'))

fig.update_layout(
    title='Total Flights per Year',
    title_font_size=20,
    xaxis_title='Year',
    yaxis_title='Total Flights (in Million)',
    xaxis_tickangle=-45,
    xaxis_showgrid=True,
    yaxis_showgrid=True
)
fig.show()
```

```python
# Define the subplot titles and colors
titles = ['BoxPlot For usg_wac', 'BoxPlot For fg_wac', 'BoxPlot For Charter', 'BoxPlot For Total']
colors = ['red', 'green', 'black', 'yellow']
columns = ['usg_wac', 'fg_wac', 'Charter', 'Total']

# Create a subplot grid
fig = make_subplots(rows=2, cols=2, subplot_titles=titles)

# Add box plots to the subplots
for i, column in enumerate(columns):
    row = i // 2 + 1
    col = i % 2 + 1
    fig.add_trace(
        go.Box(y=full_data[column], name=column, marker_color=colors[i]),
        row=row, col=col
    )

# Update layout
fig.update_layout(
    height=800, width=1000,
    title_text="BoxPlots for Different Categories",
    title_font_size=20,
    showlegend=False
)

fig.update_xaxes(title_text="Value", row=1, col=1)
fig.update_xaxes(title_text="Value", row=1, col=2)
fig.update_xaxes(title_text="Value", row=2, col=1)
fig.update_xaxes(title_text="Value", row=2, col=2)

fig.update_layout(
    title={'text': 'BoxPlot for Different Categories', 'x': 0.5, 'xanchor': 'center'},
    xaxis_title='Value',
    yaxis_title='Value'
)

fig.show()
```

```python
# Compute the correlation matrix
numerical_data = full_data.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()

# Create the heatmap
fig = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.columns,
    colorscale='Blues',
    colorbar_title='Correlation',
    zmid=0
))

# Update layout
fig.update_layout(
    title='Correlation Matrix',
    xaxis_title='Variables',
    yaxis_title='Variables',
    xaxis_nticks=len(correlation_matrix.columns),
    yaxis_nticks=len(correlation_matrix.columns),
    xaxis_tickangle=-45
)

fig.show()
```
