```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./datasets/Gold_Price_financial.csv", index_col='date', parse_dates=True)
```

```python
df = data.copy()
df
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(14, 40))
for i, col in enumerate(df.columns, 1):
    plt.subplot(16, 3, i)
    plt.plot(df[col])
    plt.title(col)
    plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 40))
for i, col in enumerate(df.columns, 1):
    plt.subplot(16, 3, i)
    sns.histplot(df[col], bins=10, kde=True)
    plt.title(col)
    plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 40))
for i, col in enumerate(df.columns, 1):
    plt.subplot(16, 3, i)
    df[col].plot(title=col, legend=True)
    plt.grid()
plt.tight_layout()
plt.show()
```
