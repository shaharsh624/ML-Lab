```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./datasets/master_pima-indians-diabetes.csv', )
```

```python
df
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
X
```

```python
y
```

```python
from sklearn.feature_selection import VarianceThreshold
var_thresh = VarianceThreshold(threshold=0.5)
var_thresh.fit(X, y)
```

```python
var_thresh.get_support()
```

```python
X[X.columns[var_thresh.get_support()]].head()
```

```python
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif
kbest = SelectKBest(score_func=chi2, k=5)
kbest_fit = kbest.fit_transform(X, y)
```

```python
kbest_cols = X.columns[kbest.get_support()]
X_kbest = df[kbest_cols]
X_kbest
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
rfe = RFE(log_reg, n_features_to_select=5)
rfe_fit = rfe.fit_transform(X, y)
```

```python
rfe_cols = X.columns[rfe.get_support()]
X_rfe = df[rfe_cols]
X_rfe
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.5)
lasso.fit(X, y)
```

```python
lasso_cols = [col for col, coef in zip(X.columns, lasso.coef_) if coef!=0]
X_lasso = df[lasso_cols]
X_lasso
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.5)
ridge.fit(X, y)
```

```python
ridge_cols = [col for col, coef in zip(X.columns, ridge.coef_) if coef>0]
X_ridge = df[ridge_cols]
X_ridge
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=8, random_state=42)
rf_classifier.fit(X, y)
```

```python
for col, coef in zip(X.columns, rf_classifier.feature_importances_): print(col, ':', coef)
```

```python
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier(eval_metric='logloss')
xgb_classifier.fit(X, y)
```

```python
for col, coef in zip(X.columns, xgb_classifier.feature_importances_): print(col, ':', coef)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# XGB Classifier without Feature Selection
xgb_original = XGBClassifier(eval_metric='logloss')
xgb_original.fit(X_train, y_train)
```

```python
y_xgb_predict_original = xgb_original.predict(X_test)
print("Original XGB Classifier")
print("Accuracy:", accuracy_score(y_test, y_xgb_predict_original))
print("F1 Score:", f1_score(y_test, y_xgb_predict_original))
print("Precision:", precision_score(y_test, y_xgb_predict_original))
print("Recall:", recall_score(y_test, y_xgb_predict_original))
```

```python
# XGB Classifier with Feature Selection
rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X_train, y_train)
```

```python
rf_sorted_indices = rf.feature_importances_.argsort()[::-1]
top_n = 5
rf_selected = np.array(X.columns)[rf_sorted_indices][:top_n]
rf_selected
```

```python
X_rf_train = X_train[rf_selected]
X_rf_test = X_test[rf_selected]
```

```python
xgb_selected = XGBClassifier(eval_metric='logloss')
xgb_selected.fit(X_rf_train, y_train)
```

```python
y_xgb_predict_selected = xgb_selected.predict(X_rf_test)
print("Selected XGB Classifier")
print("Accuracy:", accuracy_score(y_test, y_xgb_predict_selected))
print("F1 Score:", f1_score(y_test, y_xgb_predict_selected))
print("Precision:", precision_score(y_test, y_xgb_predict_selected))
print("Recall:", recall_score(y_test, y_xgb_predict_selected))
```
